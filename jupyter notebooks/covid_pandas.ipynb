```python
covid_age = pd.read_csv('age_covid.csv', index_col =0) #using pandas to read the file
covid_age.head(10) #reading the first 10 rows present in the dataset
#the index col just made covid_cases as the first column
```

```python
fig, ax = plt.subplots()

ax.bar(covid_age['Age_groups'], covid_age.index)

ax.set_title("Number of COVID-19 cases in BC in different age groups")
ax.set_ylabel('Number of COVID-19 cases')
ax.set_xlabel('Different age groups in BC')


plt.savefig('age_covid.jpg')
plt.show()
```

```python
covid_gender = pd.read_csv('gender_csv.csv', index_col =0) #using pandas to read the file
covid_gender.head(10) #reading the first 10 rows present in the dataset
#the index col just made covid_cases as the first column
```

```python
fig, ax = plt.subplots()

ax.bar(covid_gender['Gender'], covid_gender.index)

ax.set_title("Number of COVID-19 cases in BC according to gender")
ax.set_ylabel('Number of COVID-19 cases')
ax.set_xlabel('Gender')

plt.savefig('gender_covid.jpg')
plt.show()
```

```python
covid_region = pd.read_csv('datasets/region_csv.csv', index_col =0) #using pandas to read the file
covid_region.head(10) #reading the first 10 rows present in the dataset
#the index col just made covid_cases as the first column
```
